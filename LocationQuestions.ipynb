# Which key cities and cinemas have steady weekly box office, and which are more up-and-down (inconsistent) and how should Film Viet rank these locations as safer vs higher-risk targets in the future?

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sales.shape, sales.columns.tolist()
```

```python
sales = sales.copy()

sales['week_start'] = (
    sales['actual_sales_date']
      .dt.to_period('W')
      .apply(lambda r: r.start_time)
      .dt.date
)

sales[['actual_sales_date', 'week_start']].head(10)
```

```python
weekly_cinema = (
    sales
      .groupby(['state', 'city', 'theatre_name', 'week_start'], as_index=False)['gross_today']
      .sum()
      .rename(columns={'gross_today': 'weekly_gross'})
)

weekly_cinema.head()
```

```python
weekly_city = (
    sales
      .groupby(['state', 'city', 'week_start'], as_index=False)['gross_today']
      .sum()
      .rename(columns={'gross_today': 'weekly_gross'})
)

weekly_city.head()
```

```python
weekly_cinema.shape, weekly_city.shape
```

```python
cinema_stats = (
    weekly_cinema
      .groupby(['state', 'city', 'theatre_name'], as_index=False)
      .agg(
          mean_weekly_gross=('weekly_gross', 'mean'),
          std_weekly_gross=('weekly_gross', 'std'),
          weeks_count=('weekly_gross', 'count')
      )
)

cinema_stats['cv'] = cinema_stats['std_weekly_gross'] / cinema_stats['mean_weekly_gross']

cinema_stats.head()
```

```python
city_stats = (
    weekly_city
      .groupby(['state', 'city'], as_index=False)
      .agg(
          mean_weekly_gross=('weekly_gross', 'mean'),
          std_weekly_gross=('weekly_gross', 'std'),
          weeks_count=('weekly_gross', 'count')
      )
)

city_stats['cv'] = city_stats['std_weekly_gross'] / city_stats['mean_weekly_gross']

city_stats.head()
```

```python
cinema_stats[['mean_weekly_gross', 'cv', 'weeks_count']].describe()
```

```python
city_stats[['mean_weekly_gross', 'cv', 'weeks_count']].describe()
```

```python
MIN_WEEKS = 8
```

```python
cinema_stats_f = cinema_stats.copy()

cinema_stats_f = cinema_stats_f[
    (cinema_stats_f['weeks_count'] >= MIN_WEEKS) &
    (cinema_stats_f['mean_weekly_gross'] > 0) &
    (cinema_stats_f['cv'].notna())
].copy()

cinema_stats_f.shape, cinema_stats.shape
```

```python
city_stats_f = city_stats.copy()

city_stats_f = city_stats_f[
    (city_stats_f['weeks_count'] >= MIN_WEEKS) &
    (city_stats_f['mean_weekly_gross'] > 0) &
    (city_stats_f['cv'].notna())
].copy()

city_stats_f.shape, city_stats.shape
```

```python
cinema_stats_f['cv'].describe(), city_stats_f['cv'].describe()
```

```python
def classify_cv(cv):
    if cv < 0.75:
        return 'Safer (Stable)'
    elif cv < 1.10:
        return 'Moderate'
    elif cv < 1.50:
        return 'Higher-Risk (Volatile)'
    else:
        return 'Highly Volatile'
```

```python
cinema_stats_f['risk_category'] = cinema_stats_f['cv'].apply(classify_cv)
city_stats_f['risk_category'] = city_stats_f['cv'].apply(classify_cv)

cinema_stats_f['risk_category'].value_counts(), city_stats_f['risk_category'].value_counts()
```

```python
import numpy as np

X = np.log1p(heat_pivot)  # log scale

row_mean = X.mean(axis=1)
row_std = X.std(axis=1).replace(0, np.nan)

Z = (X.sub(row_mean, axis=0)).div(row_std, axis=0)  # z-score per row
```

```python
plt.figure(figsize=(16, 8))

# imshow expects a numeric matrix; keep NaN as blanks by filling with 0 but masking
Z_mat = Z.to_numpy()
mask = np.isnan(Z_mat)

# show masked NaNs as light color by setting them to 0 and using alpha mask
Z_plot = np.where(mask, 0, Z_mat)

im = plt.imshow(Z_plot, aspect='auto', interpolation='nearest')

# Make NaNs visually empty by overlaying transparent pixels
alpha = np.where(mask, 0.0, 1.0)
im.set_alpha(alpha)

plt.colorbar(im, label="Relative weekly performance (z-score of log1p gross)")

plt.yticks(range(len(Z.index)), Z.index, fontsize=9)

# reduce x labels clutter: label every ~4th week
weeks = list(Z.columns)
step = max(1, len(weeks)//12)
xticks = list(range(0, len(weeks), step))
plt.xticks(xticks, [str(weeks[i]) for i in xticks], rotation=45, ha='right', fontsize=9)

plt.title("Volatility Fingerprint: Weekly Gross Patterns (Top Safer vs Higher-risk Cinemas)")
plt.xlabel("Week start")
plt.ylabel("Cinema")

plt.tight_layout()
plt.show()
```

This heatmap shows week-by-week consistency: safer cinemas have steady patterns with few gaps, while higher-risk cinemas have more missing weeks and sharper ups-and-downs, meaning they are less predictable but can occasionally spike

# Which places watch Indian movies right away in the first 12 weeks, and which places take longer to build up (they watch more in week 3 and later)? And based on that, how should Film Viet decide where to release first and where to release later across different states, cities, and cinemas?

```python
# pick 1 film's json and 1 week key
j0 = json.loads(sales_raw.loc[0, "raw_json"])
week0 = sorted(j0.keys())[0]   # earliest week in that film's json
rows0 = j0[week0].get("rows", [])

print("week0:", week0)
print("rows in week0:", len(rows0))

r0 = rows0[0]
print("\nTop-level keys in a row:")
print(list(r0.keys()))

# Often location fields are nested; show nested keys if common containers exist
for k in ["cinema", "theatre", "site", "location", "venue"]:
    if k in r0 and isinstance(r0[k], dict):
        print(f"\nNested keys under r0['{k}']:")
        print(list(r0[k].keys())[:50])

# Compute weekly gross by summing boxOffice.day1..day7.today (if present)
bo = r0.get("boxOffice", {})
week_gross = 0
for i in range(1, 8):
    di = bo.get(f"day{i}", {})
    week_gross += float(di.get("today", 0) or 0)

print("\nWeekly gross computed from day1..day7 today:", week_gross)

# Quick peek at what looks like location text (search common field names)
candidate_fields = ["state", "city", "theatre_name", "theatreName", "cinemaName", "siteName", "locationName"]
print("\nDirect location-like fields found:")
for f in candidate_fields:
    if f in r0:
        print(f, "=>", r0[f])
```

```python
def weekly_gross_from_boxoffice(box_office: dict) -> float:
    total = 0.0
    if not isinstance(box_office, dict):
        return 0.0
    for i in range(1, 8):
        di = box_office.get(f"day{i}", {})
        if isinstance(di, dict):
            v = di.get("today", 0) or 0
            total += float(v)
    return total

rows_out = []

for _, r in sales_raw.iterrows():
    film_id = r["numero_film_id"]
    j = json.loads(r["raw_json"])

    for week_start, payload in j.items():
        week_rows = payload.get("rows", []) if isinstance(payload, dict) else []
        for rr in week_rows:
            rows_out.append({
                "numero_film_id": film_id,
                "week_start": pd.to_datetime(week_start),
                "state": rr.get("state"),
                "city": rr.get("city"),
                "theatre_name": rr.get("theatre"),
                "region": rr.get("region"),
                "circuit": rr.get("circuit"),
                "weekly_gross": weekly_gross_from_boxoffice(rr.get("boxOffice", {})),
            })

sales_weekly = pd.DataFrame(rows_out)

# Basic cleanup
sales_weekly["weekly_gross"] = sales_weekly["weekly_gross"].fillna(0).astype(float)
sales_weekly = sales_weekly.dropna(subset=["state", "city", "theatre_name", "week_start"])

print("sales_weekly shape:", sales_weekly.shape)
print("weekly_gross <= 0 rows:", (sales_weekly["weekly_gross"] <= 0).sum())
sales_weekly.head(10)
```

```python
# 1) Attach title to weekly sales
sales_wt = sales_weekly.merge(
    film_meta[["numero_film_id", "title"]],
    on="numero_film_id",
    how="left"
)

# 2) Clean titles for safer matching
sales_wt["title"] = sales_wt["title"].astype(str).str.strip()
indian_titles["title"] = indian_titles["title"].astype(str).str.strip()

# 3) Keep only Indian titles (match by title, like your earlier approach)
indian_only = sales_wt.merge(
    indian_titles[["title"]].drop_duplicates(),
    on="title",
    how="inner"
)

# 4) Drop rows with no revenue signal (prevents weight=0 issues later)
indian_only = indian_only[indian_only["weekly_gross"] > 0].copy()

print("sales_wt:", sales_wt.shape, "missing title %:", sales_wt["title"].isna().mean())
print("indian_only:", indian_only.shape, "unique films:", indian_only["numero_film_id"].nunique())
indian_only.head(10)
```

```python
# Ensure week_start is datetime
indian_only["week_start"] = pd.to_datetime(indian_only["week_start"])

def add_relative_week(df, group_cols):
    df = df.copy()
    first_week = (
        df.groupby(group_cols, as_index=False)["week_start"]
          .min()
          .rename(columns={"week_start": "first_week_start"})
    )
    out = df.merge(first_week, on=group_cols, how="left")
    out["rel_week"] = ((out["week_start"] - out["first_week_start"]).dt.days // 7) + 1
    return out

# Relative weeks at 3 levels
indian_cinema_rw = add_relative_week(
    indian_only,
    ["numero_film_id", "state", "city", "theatre_name"]
)

indian_city_rw = add_relative_week(
    indian_only,
    ["numero_film_id", "state", "city"]
)

indian_state_rw = add_relative_week(
    indian_only,
    ["numero_film_id", "state"]
)

# Quick check
print("cinema_rw:", indian_cinema_rw.shape, "rel_week range:", (indian_cinema_rw["rel_week"].min(), indian_cinema_rw["rel_week"].max()))
print("city_rw:", indian_city_rw.shape, "rel_week range:", (indian_city_rw["rel_week"].min(), indian_city_rw["rel_week"].max()))
print("state_rw:", indian_state_rw.shape, "rel_week range:", (indian_state_rw["rel_week"].min(), indian_state_rw["rel_week"].max()))

indian_cinema_rw[["numero_film_id","title","state","city","theatre_name","week_start","first_week_start","rel_week","weekly_gross"]].head(10)
```

```python
def build_film_location_timing(df, group_cols):
    df = df.copy()

    df["early_gross"] = np.where(df["rel_week"] <= 2, df["weekly_gross"], 0.0)
    df["late_gross"]  = np.where(df["rel_week"] >= 3, df["weekly_gross"], 0.0)

    out = (
        df.groupby(group_cols, as_index=False)
          .agg(
              total_gross=("weekly_gross", "sum"),
              early_gross=("early_gross", "sum"),
              late_gross=("late_gross", "sum"),
              weeks_active=("rel_week", "max")
          )
    )

    out["early_share"] = np.where(out["total_gross"] > 0, out["early_gross"] / out["total_gross"], np.nan)
    return out

# --- Cinema level: (film, state, city, theatre) ---
film_cinema_timing = build_film_location_timing(
    indian_cinema_rw,
    ["numero_film_id", "title", "state", "city", "theatre_name"]
)

# --- City level: (film, state, city) ---
film_city_timing = build_film_location_timing(
    indian_city_rw,
    ["numero_film_id", "title", "state", "city"]
)

# --- State level: (film, state) ---
film_state_timing = build_film_location_timing(
    indian_state_rw,
    ["numero_film_id", "title", "state"]
)

# Quick quality checks (these should look sensible)
print("film_cinema_timing:", film_cinema_timing.shape,
      "bad early_share:", ((film_cinema_timing["early_share"] < 0) | (film_cinema_timing["early_share"] > 1)).sum(),
      "min total_gross:", film_cinema_timing["total_gross"].min(),
      "weeks_active range:", (film_cinema_timing["weeks_active"].min(), film_cinema_timing["weeks_active"].max())
)

print("film_city_timing:", film_city_timing.shape,
      "bad early_share:", ((film_city_timing["early_share"] < 0) | (film_city_timing["early_share"] > 1)).sum(),
      "min total_gross:", film_city_timing["total_gross"].min(),
      "weeks_active range:", (film_city_timing["weeks_active"].min(), film_city_timing["weeks_active"].max())
)

print("film_state_timing:", film_state_timing.shape,
      "bad early_share:", ((film_state_timing["early_share"] < 0) | (film_state_timing["early_share"] > 1)).sum(),
      "min total_gross:", film_state_timing["total_gross"].min(),
      "weeks_active range:", (film_state_timing["weeks_active"].min(), film_state_timing["weeks_active"].max())
)

film_city_timing.head(10)
```

```python
def safe_weighted_avg(values, weights):
    values = pd.Series(values)
    weights = pd.Series(weights).fillna(0)
    wsum = weights.sum()
    if wsum <= 0:
        return np.nan
    return np.average(values, weights=weights)

def build_place_summary(film_location_timing, place_cols):
    df = film_location_timing[film_location_timing["total_gross"] > 0].copy()
    out = (
        df.groupby(place_cols, as_index=False)
          .apply(lambda g: pd.Series({
              "total_gross": g["total_gross"].sum(),
              "n_films": g["numero_film_id"].nunique(),
              "weighted_early_share": safe_weighted_avg(g["early_share"], g["total_gross"])
          }))
          .reset_index(drop=True)
    )
    return out

# State summary
state_summary = build_place_summary(film_state_timing, ["state"])

# City summary
city_summary = build_place_summary(film_city_timing, ["state", "city"])

# Cinema summary
cinema_summary = build_place_summary(film_cinema_timing, ["state", "city", "theatre_name"])

print("state_summary:", state_summary.shape, state_summary[["total_gross","weighted_early_share"]].isna().sum().to_dict())
print("city_summary:", city_summary.shape, city_summary[["total_gross","weighted_early_share"]].isna().sum().to_dict())
print("cinema_summary:", cinema_summary.shape, cinema_summary[["total_gross","weighted_early_share"]].isna().sum().to_dict())

city_summary.sort_values("total_gross", ascending=False).head(10)
```

```python
# 1) City-week totals across all Indian films
city_week = (
    weekly_city_rel
    .groupby(["state", "city", "rel_week"], as_index=False)["weekly_gross"]
    .sum()
)

# 2) Compute share of total gross by rel_week within each (state, city)
city_totals = (
    city_week
    .groupby(["state", "city"], as_index=False)["weekly_gross"]
    .sum()
    .rename(columns={"weekly_gross": "city_total_gross"})
)

city_week = city_week.merge(city_totals, on=["state", "city"], how="left")
city_week["week_share"] = city_week["weekly_gross"] / city_week["city_total_gross"]

# 3) Cumulative share curve (how fast a city captures revenue over weeks)
city_week = city_week.sort_values(["state", "city", "rel_week"])
city_week["cum_share"] = city_week.groupby(["state", "city"])["week_share"].cumsum()

print("city_week shape:", city_week.shape)
print("week_share min/max:", city_week["week_share"].min(), city_week["week_share"].max())
print("cum_share min/max:", city_week["cum_share"].min(), city_week["cum_share"].max())
city_week.head(12)
```

```python
# 1) How many rel_weeks exist per city?
city_weeks_count = (
    city_week
    .groupby(["state", "city"], as_index=False)
    .agg(n_weeks=("rel_week", "nunique"))
)

# 2) Weeks to hit 80% and 95% cumulative revenue
def first_week_reach(df, threshold):
    hit = df.loc[df["cum_share"] >= threshold, "rel_week"]
    return int(hit.min()) if len(hit) else np.nan

city_speed = (
    city_week
    .sort_values(["state", "city", "rel_week"])
    .groupby(["state", "city"], as_index=False)
    .apply(lambda g: pd.Series({
        "weeks_to_80": first_week_reach(g, 0.80),
        "weeks_to_95": first_week_reach(g, 0.95),
        "final_week": int(g["rel_week"].max()),
    }))
    .reset_index(drop=True)
)

# 3) Merge week counts
city_speed = city_speed.merge(city_weeks_count, on=["state", "city"], how="left")

print("city_speed shape:", city_speed.shape)
print(city_speed[["weeks_to_80","weeks_to_95","final_week","n_weeks"]].describe())
city_speed.sort_values(["weeks_to_95","weeks_to_80"]).head(12)
```
